# Consistency of Responses (Success Grading)

This notebook will look at the numerical grading of responses to evaluate consistency.

## IMPORT DATA

```python
import pandas as pd
dataframe = pd.read_csv('./Testing transcribed into CSV.csv')
dataframe
```

## CONSISTENCY (BOX & WHISKER) - Overall

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(data=dataframe, x="Success", palette="Spectral")
plt.xlabel("Success Grading")
plt.show()
```

# CONSISTENCY (BOX & WHISKER) - By file class

```python
sns.boxplot(data=dataframe, x="Success", y="File", hue="File", palette="Spectral")
plt.xlabel("Success")
plt.show()
```
