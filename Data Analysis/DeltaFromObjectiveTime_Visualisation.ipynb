# Data Visualisation (Delta from Objective Time.)
This notebook will visualise the data from the delta from objective time derived dependent variable.

## IMPORT DATA.

```python
import pandas as pd
dataframe = pd.read_csv('./delta_from_objective_time_working_data')
dataframe
```

## CONSISTENCY (BOX & WHISKER) - Overall

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(data=dataframe, x="Delta from Objective Time", palette="Spectral")
plt.xlabel("Delta from Objective Time (ms)")
plt.show()
```

## CONSISTENCY (BOX & WHISKER) - By File Class

```python
sns.boxplot(data=dataframe, x="Delta from Objective Time", y="File", hue="File", palette="Spectral")
plt.xlabel("Delta from Objective Time (ms)")
plt.show()
```

## CONSISTENCY (VIOLIN) - Overall

```python
sns.catplot(data=dataframe, x="Delta from Objective Time", kind="violin")
plt.show()
```

## CONSISTENCY (VIOLIN) - By File Class

```python
sns.catplot(data=dataframe, x="Delta from Objective Time", y="File", hue="File", kind="violin")
plt.show()
```
